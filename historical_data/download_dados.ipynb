```python
import MetaTrader5 as mt5
from datetime import datetime
import pandas as pd
import pytz
```

```python
mt5.initialize()
```

```python
timezone = pytz.timezone("Etc/UTC")
utc_from = datetime(2020, 12, 30, tzinfo=timezone)
def download_cotacao(ticker):  
    rates = mt5.copy_rates_from(ticker, mt5.TIMEFRAME_M5, utc_from, 99000) #baixa dados
    dados = pd.DataFrame(rates) #cria dataframe
    dados['time']=pd.to_datetime(dados['time'], unit='s')
    #dados = dados.set_index('time')
    #dados = dados['close'] 
    #dados = dados.rename(ticker)
    return dados
```
